```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd

df_plaquesize = pd.read_csv('MP1plaquesize.csv', header=None)*1000 #mm to micron
df_zoi = pd.read_csv('MP1ZOI.csv', header=None)*1000
```

```python
plt.rcParams.update({'font.size': 12})
size = 19
ftarr = np.linspace(1, 10, int(size/2+1))
fbarr = np.linspace(1, 10, int(size/2+1))
origin = "lower"
vmin = min([np.min(df_zoi.to_numpy()/1000),np.min(df_plaquesize.to_numpy()/1000)])
vmax = max([np.max(df_zoi.to_numpy()/1000),np.max(df_plaquesize.to_numpy()/1000)])
#vmin = 0.82

plt.imshow(df_zoi.T/1000, origin=origin,        cmap="plasma", vmin=vmin, vmax=vmax)
plt.colorbar()
plt.xticks(np.arange(size, step=2), ftarr,fontsize = 11)
plt.yticks(np.arange(size, step=2), fbarr,fontsize = 11)
plt.xlabel(r"$f_{\tau}$",fontsize = 14)
plt.ylabel(r"$f_{\beta}$",fontsize = 14)
plt.savefig("Fig5B.png")
plt.figure()
plt.imshow(df_plaquesize.T/1000, origin=origin, cmap="plasma", vmin=vmin, vmax=vmax)
plt.colorbar()
plt.xticks(np.arange(size, step=2), ftarr,fontsize = 11)
plt.yticks(np.arange(size, step=2), fbarr,fontsize = 11)
plt.xlabel(r"$f_{\tau}$",fontsize = 14)
plt.ylabel(r"$f_{\beta}$",fontsize = 14)
plt.savefig("Fig5A.png")
plt.show()
```
