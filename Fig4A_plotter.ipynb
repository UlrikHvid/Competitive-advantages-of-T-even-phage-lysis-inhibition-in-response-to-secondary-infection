```python
import pandas as pd

df_plaquesize = pd.read_csv('MP1plaquesize.csv', header=None)*1000 #mm to micron
df_zoi = pd.read_csv('MP1ZOI.csv', header=None)*1000
```

```python
#Values for MP0 (no LIN)
rhalf0 = 1192
rZOI0  = 1306
```

```python
size = 4
ftarr = np.linspace(1, 10, size)
fbarr = np.linspace(1, 10, size)
origin = "lower"
plt.imshow(df_zoi.T/rZOI0, origin = origin)
plt.colorbar()
plt.xticks([0,1,2,3],ftarr)
plt.yticks([0,1,2,3],fbarr)
plt.xlabel(r"$f_{\tau}$" )
plt.ylabel(r"$f_{\beta}$")
plt.figure()
plt.imshow(df_plaquesize.T/rhalf0, origin = origin)
plt.colorbar()
plt.xticks([0,1,2,3],ftarr)
plt.yticks([0,1,2,3],fbarr)
plt.xlabel(r"$f_{\tau}$" )
plt.ylabel(r"$f_{\beta}$")
plt.show()
```
